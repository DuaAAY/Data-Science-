Groupby

```python
import numpy as np
import pandas as pd
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'], 
        'Person':['Sam', 'Charlie','Amy','Vanessa','Carl','Sarah'], 
        'Sales':[200,120,340,124,243,350]}
```

```python
tc = pd.DataFrame(data)
```

```python
tc
```

```python
comp = tc.groupby('Company')
```

```python
comp.mean()
```

```python
comp.std()
```

```python
comp = tc.groupby('Company').sum().loc['FB'] #all in one line
```

```python
tc.groupby('Person').count().loc['Sam'] #count works with string
```

```python
tc.groupby('Company').describe().transpose()
```

2- Merging

```python
#when you merge 2 data frames using thier mutual cloumn like key cloumn in the below example
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})  
```

```python
left
```

```python
right
```

```python
pd.merge(left,right, how='inner', on = 'key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left,right, on= ['key1', 'key2']) #by defult it will merge on inner
```

```python
pd.merge(left,right, how = 'inner', on = ['key1','key2'])
```

```python
pd.merge(left,right, how = 'outer', on = ['key1','key2'])
```

```python
pd.merge(left,right, how = 'left', on = ['key1','key2'])
```

```python
pd.merge(left,right, how = 'right', on = ['key1','key2'])
```

3- Join

```python
#when you want to join 2 data frames having the diffrent index 
#join of index instead of cloumn
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```

```python
right.join(left, how='outer')
```
