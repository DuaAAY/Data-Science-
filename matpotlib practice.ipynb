Figure size and DPI

```python
f = np.arange(0,11)
c = f**2
```

```python
f
```

```python
c
```

```python
figr = plt.figure(figsize=(8,2))
g = figr.add_axes([0,0,1,1])
g.plot(f,c)
```

```python
figr,g = plt.subplots(nrows=2,ncols= 1, figsize=(8,2))
g[0].plot(f,c)
g[0].set_title('Increasing')
g[1].plot(c,f)
plt.tight_layout()
```

```python
figr.savefig('my_picture.png', dpi= 200) #pixels
```

```python
figo = plt.figure()
cx = figo.add_axes([0,0,1,1])
cx.plot(f,c)
cx.plot(c,f)
cx.set_title("Dua's chart")
cx.set_xlabel('x')
cx.set_ylabel('y')
```

```python
figo = plt.figure()
cx = figo.add_axes([0,0,1,1])
cx.plot(f,f**2, label = 'value squred')
cx.plot(f,f**3, label= 'value cubed')
cx.set_xlabel('x')
cx.set_ylabel('y')
cx.legend()
```
